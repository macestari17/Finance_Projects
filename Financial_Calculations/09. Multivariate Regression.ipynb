# Import Data

```python
import pandas_datareader.data as web

ff_data = web.DataReader("F-F_Research_Data_Factors", "famafrench")

# Extract the main dataset into a DataFrame (Access the data in dictionary stored under key 0)
ff_data_frame = ff_data[0]

ff_data_frame
```

```python
ff_data_frame.info()
```

```python
start_date = ff_data_frame.index[0].strftime('%Y-%m-%d')
end_date = ff_data_frame.index[-1].strftime('%Y-%m-%d')
start_date, end_date
```

```python
import yfinance as yf
import pandas as pd

tickers = ['PG', '^GSPC']

reg_data = pd.DataFrame()

for t in tickers:      
    reg_data[t] = yf.download(t, start=start_date, end=end_date, interval='1mo')['Adj Close']
```

```python
reg_data
```

```python
# Convert index to 'YYYY-MM'
reg_data.index = reg_data.index.to_period('M')
reg_data
```

# Calculate Simple Return

```python
reg_returns = (reg_data / reg_data.shift(1))-1
```

```python
reg_returns
```

```python
reg_returns.info()
```

```python
# Right=first df, left=second df
# Outer=ensures that all rows from both DataFrames are included (rows with missing values in one DataFrame will have NaN
data = pd.merge(reg_returns, ff_data_frame, left_index=True, right_index=True, how='outer')
data = data.dropna()
data
```

```python
data.info()
```

```python
X = data [['^GSPC', 'SMB', 'HML', 'RF']]
Y = data ['PG']
```

```python
import numpy as np

print(X.isnull().sum())
print(Y.isnull().sum())
print(np.isinf(X).sum())
print(np.isinf(Y).sum()) 
```

# Create Regression

```python
import statsmodels.api as sm

# Assign a constant (newly obtained information) as X1
X1 = sm.add_constant(X)

# Assign Ordinary Least Squared Regression (OLS) as reg
reg = sm.OLS(Y, X1).fit()
```

```python
reg.summary()
```
