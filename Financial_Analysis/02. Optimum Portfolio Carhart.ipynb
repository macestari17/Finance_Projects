# Linear Regression

```python
# List of factors to exclude
carhart_factors = ['^GSPC', 'SMB', 'HML', 'RF']

# Redefine tickers: include all columns except the Carhart factors
tickers = [
    col for col in data.columns
    if col not in carhart_factors
]
```

```python
from scipy import stats
import statsmodels.api as sm

# Initialize an empty dictionary to store the regression results
regression = {}

# Define X as the returns of Carhart factors
X = data[['^GSPC', 'SMB', 'HML', 'RF']]  # Use a list to select multiple columns

# Add a constant for the intercept in statsmodels
X_with_const = sm.add_constant(X)

# Loop through each ticker in the tickers list
for ticker in tickers:
    Y = data[ticker]  # Get the returns of the current ticker

    # Perform regression with statsmodels to get the p-value for the intercept
    model = sm.OLS(Y, X_with_const).fit()
    slope = model.params['^GSPC']  # Slope for the market factor
    p_value_slope = model.pvalues['^GSPC']  # P-value for the market factor
    p_value_intercept = model.pvalues['const']  # P-value for intercept

    # Apply transformations
    alpha = model.params['const'] * 12  # Annualize alpha
    r_square = model.rsquared  # R-squared value

    # Store the transformed results in the dictionary
    regression[ticker] = {
        'slope': slope,
        'alpha': alpha,
        'r_square': r_square,
        'p_value_slope': p_value_slope,  # P-value for slope
        'p_value_intercept': p_value_intercept,  # P-value for intercept
        'std_err': model.bse['const'],  # Standard error for intercept
    }
```

### Alphas

```python
# Filter the dictionary by values where p-value <= 0.1 and alpha > 0
# Initialize an empty dictionary to store filtered results
filtered_regression = {}

# Loop through each ticker and its regression values
for ticker, values in regression.items():
    p_value_intercept = values['p_value_intercept']  # Access the p-value for the constant term
    alpha = values['alpha']  # Access the alpha value
    
    # Check if the p-value is <= 0.1 and alpha > 0
    if p_value_intercept <= 0.1 and alpha > 0:
        filtered_regression[ticker] = values  # Add the ticker to the filtered dictionary
```

### Variances, Betas, Systematic and Unsystematic Risks

```python
from scipy import stats

# Initialize the dictionary to store results
statistics = {}

# Loop through alpha tickers
for ticker in alpha_tickers:
    var = excess_returns[ticker].var() * 12  # Annual variance
    beta = sorted_regression_alpha[ticker]['slope']  # Beta
    sys_risk = beta**2 * SP500_var  # Systematic risk
    unsys_risk = var - sys_risk  # Unsystematic risk

    # Store results with labeled keys
    statistics[ticker] = {
        'Variance': var,
        'Beta': beta,
        'Systematic Risk': sys_risk,
        'Unsystematic Risk': unsys_risk
    }

# Display the results
statistics
```

```python
# Convert the statistics dictionary to a DataFrame
statistics_df = pd.DataFrame.from_dict(statistics, orient='index').T

# Display the resulting DataFrame
statistics_df
```

# Active Portfolio

```python
# Extract alpha values (item[1][1]) from top_20_alpha
alpha_values = {item[0]: item[1]['alpha'] for item in top_20_alpha}

# Initialize an empty dictionary
adjusted_alpha_values = {}

# Loop through each key-value pair in alpha_values to adjust apha at 2%
for key, value in alpha_values.items():
    # Multiply the value by 0.1 to keep 10%
    adjusted_value = value * 0.1
    
    # Add the key and adjusted value to the new dictionary
    adjusted_alpha_values[key] = adjusted_value

# Convert adjusted alpha values to a DataFrame row
alpha_row = pd.DataFrame([adjusted_alpha_values], index=["Alpha"])

# Concatenate the adjusted alpha row to statistics_df
statistics_df = pd.concat([statistics_df, alpha_row])

statistics_df
```

```python
# Calculate Reward to Risk Ratio
statistics_df.loc["Reward to Risk Ratio"] = statistics_df.loc["Alpha"] / statistics_df.loc["Unsystematic Risk"]
reward_to_risk_total = statistics_df.loc["Reward to Risk Ratio"].sum()
statistics_df.loc["Reward to Risk Ratio", "Total"] = reward_to_risk_total

# Calculate Active Weightings
statistics_df.loc["Active Weightings"] = statistics_df.loc["Reward to Risk Ratio"] / reward_to_risk_total

# Calculate Active Portfolio A
statistics_df.loc["Active Portfolio A"] = statistics_df.loc["Alpha"] * statistics_df.loc["Active Weightings"]
active_portfolio_total = statistics_df.loc["Active Portfolio A"].sum()
statistics_df.loc["Active Portfolio A", "Total"] = active_portfolio_total

# Calculate Portfolio Unsyst. Risk
statistics_df.loc["Portfolio Unsyst. Risk"] = statistics_df.loc["Unsystematic Risk"] * statistics_df.loc["Active Weightings"] ** 2
portfolio_unsyst_risk_total = statistics_df.loc["Portfolio Unsyst. Risk"].sum()
statistics_df.loc["Portfolio Unsyst. Risk", "Total"] = portfolio_unsyst_risk_total

# Calculate Active Portfolio Reward to Risk (A/A)
active_portfolio_reward_to_risk = active_portfolio_total / portfolio_unsyst_risk_total
statistics_df.loc["Active Portfolio Reward to Risk (A/A)", "Total"] = active_portfolio_reward_to_risk

# Calculate Annual information ratio
statistics_df.loc["Annual Information Ratio"] = statistics_df.loc["Active Portfolio A"] / statistics_df.loc["Portfolio Unsyst. Risk"] ** 0.5

# Display the final DataFrame
statistics_df
```

### Active Portfolio A

```python
active_portfolio_beta = (statistics_df.loc["Beta"] * statistics_df.loc["Active Weightings"]).sum()
active_portfolio_beta
```

### Market Reward to Risk

```python
market_reward_to_risk = market_risk_premium / SP500_var
market_reward_to_risk
```

### Weighting in Active Portfolio w0A

```python
weighting_in_active_portfolio = statistics_df.loc["Active Portfolio Reward to Risk (A/A)", "Total"] / market_reward_to_risk
weighting_in_active_portfolio
```

### Weight in Active Portfolio w*A (adjusting for beta)

```python
weighting_in_active_portfolio = weighting_in_active_portfolio / (1+(1 - active_portfolio_beta) * weighting_in_active_portfolio)
weighting_in_active_portfolio
```

### Weight in S&P500

```python
weighting_in_SP500 = 1 - weighting_in_active_portfolio
weighting_in_SP500
```

### Weights Without Leverage

```python
# Calculate the total sum of absolute weights
total_abs_weights = abs(weighting_in_SP500) + abs(weighting_in_active_portfolio)

# Normalize each weight
weighting_in_SP500_without_leverage = weighting_in_SP500 / total_abs_weights
weighting_in_active_portfolio_without_leverage = weighting_in_active_portfolio / total_abs_weights

# Print results
print("weighting_in_SP500_without_leverage:", weighting_in_SP500_without_leverage)
print("weighting_in_active_portfolio_without_leverage:", weighting_in_active_portfolio_without_leverage)
```

```python
revised_sharpe_ratio = (SP500_sharpe_ratio  ** 2 + active_portfolio_information_ratio ** 2) ** 0.5
revised_sharpe_ratio = revised_sharpe_ratio.round(7)
revised_sharpe_ratio
```

### Sharpe Ratio Improvement

```python
sharpe_ratio_improvement = revised_sharpe_ratio - SP500_sharpe_ratio
sharpe_ratio_improvement = sharpe_ratio_improvement.round(7)
sharpe_ratio_improvement
```

# Verify Results

### Portfolio (P*) Risk Premium

```python
portfolio_risk_premium = weighting_in_SP500 * market_risk_premium + weighting_in_active_portfolio * (active_portfolio_beta * market_risk_premium + active_portfolio_total)
portfolio_risk_premium
```

### Active Portfolio Standard Deviation

```python
active_portfolio_std = (((active_portfolio_beta ** 2) * SP500_var) + portfolio_unsyst_risk_total) ** 0.5
active_portfolio_std
```

### Active Portfolio Covariance with M

```python
active_portfolio_cov_with_market = active_portfolio_beta * SP500_var
active_portfolio_cov_with_market
```

### Portfolio (P*) Standard Deviation

```python
portfolio_std = ((weighting_in_SP500 ** 2) * SP500_var + (weighting_in_active_portfolio ** 2) * (active_portfolio_std ** 2) + (2 * weighting_in_active_portfolio) * weighting_in_SP500 * active_portfolio_cov_with_market) ** 0.5
portfolio_std
```

### Portfolio (P*) Sharpe Ratio

```python
portfolio_sharpe_ratio = portfolio_risk_premium / portfolio_std
portfolio_sharpe_ratio = portfolio_sharpe_ratio.round(7)
print (portfolio_sharpe_ratio)
portfolio_sharpe_ratio > SP500_sharpe_ratio and portfolio_sharpe_ratio == revised_sharpe_ratio
```
