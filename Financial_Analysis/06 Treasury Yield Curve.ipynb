```python
import yfinance as yf
from datetime import datetime, timedelta

# Define the tickers for Treasury rates
tickers = [
    "^IRX",
    "2YY=F",
    "^FVX",
    "^TNX",
    "^TYX",
]

treasury_yield_today = {}

# Download data for the past 5 days (in case it's not a trading day)
start_date = (datetime.today() - timedelta(days=5)).strftime('%Y-%m-%d')
end_date = datetime.today().strftime('%Y-%m-%d')

# Download rates
for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    treasury_yield_today[ticker] = data['Close'].iloc[-1]  # Get the most recent rate

treasury_yield_today
```

```python
import matplotlib.pyplot as plt

# Define the tenors (labels for the x-axis)
tenors = ["13-Week", "2-Year", "5-Year", "10-Year", "30-Year"]

# Get the corresponding rates from the dictionary
rates = [treasury_yield_today[ticker] for ticker in tickers]

# Plot the yield curve
plt.figure(figsize=(10, 6))
plt.plot(tenors, rates, marker='o', label="Yield Curve Today", color='blue')

# Add labels and title
plt.title("U.S. Treasury Yield Curve", fontsize=16)
plt.xlabel("Tenor", fontsize=14)
plt.ylabel("Interest Rate (%)", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the tenors (labels for the x-axis)
tenors = ["13-Week", "2-Year", "5-Year", "10-Year", "30-Year"]

# Define the yield data dictionaries
yield_curves = {
    "Today": treasury_yield_today,
    "30 Days Ago": treasury_yield_30_days_ago,
    "90 Days Ago": treasury_yield_90_days_ago,
    "180 Days Ago": treasury_yield_180_days_ago,
    "1 Year Ago": treasury_yield_1_year_ago,
    "2 Years Ago": treasury_yield_2_years_ago
}

# Generate a colormap for lighter to darker blue shades
colors = plt.cm.Blues(np.linspace(0.9, 0.3, len(yield_curves)))  # Reverse the gradient

# Plot each yield curve
plt.figure(figsize=(12, 8))
for (time_period, yields), color in zip(yield_curves.items(), colors):
    rates = [yields[ticker] for ticker in tickers]  # Extract rates for the tenors
    plt.plot(tenors, rates, marker='o', label=time_period, color=color)

# Add labels, title, and legend
plt.title("U.S. Treasury Yield Curves Over Time", fontsize=16)
plt.xlabel("Tenor", fontsize=14)
plt.ylabel("Interest Rate (%)", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="Time Period", fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Define the tenors (labels for the x-axis)
tenors = ["13-Week", "2-Year", "5-Year", "10-Year", "30-Year"]

# Define the yield data dictionaries
yield_curves = {
    "Today": treasury_yield_today,
    "30 Days Ago": treasury_yield_30_days_ago,
    "90 Days Ago": treasury_yield_90_days_ago,
    "180 Days Ago": treasury_yield_180_days_ago,
    "1 Year Ago": treasury_yield_1_year_ago,
    "2 Years Ago": treasury_yield_2_years_ago
}

# Generate a colormap for lighter to darker blue shades
colors = plt.cm.Blues(np.linspace(0.9, 0.3, len(yield_curves)))  # Reverse the gradient

# Define markers for each time period
markers = ['o', 's', '^', 'D', 'v', '*']  # Add more markers if needed

plt.figure(figsize=(12, 8))
for (time_period, yields), color, marker in zip(yield_curves.items(), colors, markers):
    linewidth = 2
    plt.plot(
        tenors,
        [yields[ticker] for ticker in tickers],  # Extract values for the tenors
        marker=marker,
        label=time_period,
        color=color,
        linewidth=linewidth
    )

# Add title and labels
plt.title("U.S. Treasury Yield Curves Over Time", fontsize=18)
plt.xlabel("Tenor", fontsize=14)
plt.ylabel("Interest Rate (%)", fontsize=14)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.8, color='gray')

# Adjust legend
plt.legend(title="Time Period", fontsize=12)

# Tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
