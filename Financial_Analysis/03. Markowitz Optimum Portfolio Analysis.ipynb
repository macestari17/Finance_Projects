# Linear Regression

```python
# List of factors to exclude
carhart_factors = ['^GSPC', 'SMB', 'HML', 'RF']

# Redefine tickers: include all columns except the Carhart factors
tickers = [
    col for col in data.columns
    if col not in carhart_factors
]
```

```python
from scipy import stats
import statsmodels.api as sm

# Initialize an empty dictionary to store the regression results
regression = {}

# Define X as the returns of Carhart factors
X = data[['^GSPC', 'SMB', 'HML', 'RF']]  # Use a list to select multiple columns

# Add a constant for the intercept in statsmodels
X_with_const = sm.add_constant(X)

# Loop through each ticker in the tickers list
for ticker in tickers:
    Y = data[ticker]  # Get the returns of the current ticker

    # Perform regression with statsmodels to get the p-value for the intercept
    model = sm.OLS(Y, X_with_const).fit()
    slope = model.params['^GSPC']  # Slope for the market factor
    p_value_slope = model.pvalues['^GSPC']  # P-value for the market factor
    p_value_intercept = model.pvalues['const']  # P-value for intercept

    # Apply transformations
    alpha = model.params['const'] * 12  # Annualize alpha
    r_square = model.rsquared  # R-squared value

    # Store the transformed results in the dictionary
    regression[ticker] = {
        'slope': slope,
        'alpha': alpha,
        'r_square': r_square,
        'p_value_slope': p_value_slope,  # P-value for slope
        'p_value_intercept': p_value_intercept,  # P-value for intercept
        'std_err': model.bse['const'],  # Standard error for intercept
    }
```

### Alphas

```python
# Filter the dictionary by values where p-value <= 0.1 and alpha > 0
# Initialize an empty dictionary to store filtered results
filtered_regression = {}

# Loop through each ticker and its regression values
for ticker, values in regression.items():
    p_value_intercept = values['p_value_intercept']  # Access the p-value for the constant term
    alpha = values['alpha']  # Access the alpha value
    
    # Check if the p-value is <= 0.1 and alpha > 0
    if p_value_intercept <= 0.1 and alpha > 0:
        filtered_regression[ticker] = values  # Add the ticker to the filtered dictionary
```

### Bounds and Initial Weights Parameters (for all Solve Calculations)

```python
import numpy as np
# Annualized Returns excluding S&P500
returns = basic_stats_df.loc[basic_stats_df.index != 'S&P500', "Annualised Returns"].values
```

```python
# Monthly returns matrix (from your data)
monthly_returns = alpha_returns.drop(columns='S&P500').values  # Drop the 'S&P500' column
```

```python
# Number of assets
n_assets = len(returns)
n_assets
```

```python
# BOUNDS: Define the range for each weight (uncomment the bounds you want to use:)
# 1. (0, 1): No short selling, weights are between 0 and 1
# 2. (-1, 1): Allow short selling, weights can be negative (represents borrowing or selling assets not owned)
# 3. (None, None): No restriction on weights, allows leverage (weights can exceed 1) and short selling (negative weights)

bounds = [(0, 1) for _ in range(n_assets)]  # No short selling
# bounds = [(-1, 1) for _ in range(n_assets)]  # Allow short selling
# bounds = [(None, None) for _ in range(n_assets)]  # Allow weights to be greater than 1 (100%) (default in Excel Solver)
```

```python
# Initial guess for INITIAL WEIGHTS (zeros or equal allocation) (uncomment the initialization method you want to use)
# 1. Equal allocation: Start with equal weights across all assets
# 2. All weights set to 1: Useful when leveraging is allowed and starting from a fully allocated portfolio
# 3. All weights set to 0: Start with no allocation (useful in specific scenarios or for comparison)

# initial_weights = np.ones(n_assets) / n_assets  # Start with equal allocation
# initial_weights = np.ones(n_assets) # Starts with 1s
initial_weights = np.zeros(n_assets)  # Start with zeros
```

```python
from scipy.optimize import minimize
import pandas as pd

# Constraint: Sum of weights must equal 1
constraints = [
    {'type': 'eq', 'fun': lambda w: np.sum(w) - 1},  # Sum of weights = 1
    {'type': 'eq', 'fun': lambda w: np.dot(w, returns) - target_return}  # Portfolio return = Target return
]

# Objective function: Portfolio variance (to minimize)
def portfolio_variance(weights):
    weighted_returns = np.dot(monthly_returns, weights)  # Weighted monthly returns
    monthly_variance = np.var(weighted_returns)  # Variance of weighted returns
    annual_variance = monthly_variance * 12  # Annualize the variance
    return annual_variance

# Optimize the portfolio
result = minimize(portfolio_variance, initial_weights, bounds=bounds, constraints=constraints)

# Optimal weights
optimal_weights = result.x

# Calculate the portfolio variance using the optimal weights
optimal_variance = portfolio_variance(optimal_weights)

# Create a DataFrame to show the weights with asset names
asset_names = basic_stats_df.loc[basic_stats_df.index != 'S&P500'].index  # Exclude S&P500
weights_df = pd.DataFrame({'Asset': asset_names, 'Weight': optimal_weights})
weights_df.set_index('Asset', inplace=True)

# Print the results
print(weights_df)
print(f"Portfolio Variance (Annualized): {optimal_variance:.6f}")
```

# Efficient Frontier

### Minimum Variance Frontier

```python
# Generate random target returns
random_target_returns = np.linspace(0, 0.5, 100)  # Generate 1000 evenly spaced target returns in range 0 to 0.5

# Objective function: Portfolio variance (to minimize)
def portfolio_variance(weights):
    weighted_returns = np.dot(monthly_returns, weights)  # Weighted monthly returns
    monthly_variance = np.var(weighted_returns)  # Variance of weighted returns
    annual_variance = monthly_variance * 12  # Annualize the variance
    return annual_variance

# Store results
results = []

for target_return in random_target_returns:
    # Constraints for optimization
    constraints = [
        {'type': 'eq', 'fun': lambda w: np.sum(w) - 1},  # Sum of weights = 1
        {'type': 'eq', 'fun': lambda w: np.dot(w, returns) - target_return}  # Portfolio return = Target return
    ]
    
    # Optimize
    result = minimize(portfolio_variance, initial_weights, bounds=bounds, constraints=constraints)
    
    # Append results directly
    optimal_weights = result.x
    optimal_variance = portfolio_variance(optimal_weights)
    results.append({
        'Target Return': target_return,
        'Portfolio Standard Deviation': optimal_variance ** 0.5,
        'Weights': optimal_weights
    })

# Create a DataFrame to store the results
results_df = pd.DataFrame(results)

# Expand weights into individual columns for readability
weights_expanded = pd.DataFrame(results_df['Weights'].tolist(), columns=basic_stats_df.loc[basic_stats_df.index != 'S&P500'].index)
results_df = pd.concat([results_df.drop(columns='Weights'), weights_expanded], axis=1)

results_df = results_df.sort_values(by='Target Return', ascending=True)
results_df
```

```python
import matplotlib.pyplot as plt

# Plot the Minimum Variance Frontier
plt.figure(figsize=(10, 6))
plt.plot(
    results_df['Portfolio Standard Deviation'],  # X-axis: Standard Deviation (Risk)
    results_df['Target Return'],                # Y-axis: Target Return (Expected Return)
    label='Minimum Variance Frontier',
)
plt.xlabel('Portfolio Standard Deviation')
plt.ylabel('Expected Return')
plt.title('Minimum Variance Frontier')
plt.grid(True)
plt.legend()
plt.show()
```

```python
portfolio_sharpe_ratios_df = results_df
portfolio_sharpe_ratios_df['Sharpe Ratios'] = (results_df['Target Return'] - risk_free) / results_df['Portfolio Standard Deviation']

# Move 'Sharpe Ratios' to the first position
portfolio_sharpe_ratios_df.insert(2, 'Sharpe Ratios', portfolio_sharpe_ratios_df.pop('Sharpe Ratios'))

portfolio_sharpe_ratios_df
```

### Global Minimum Efficient Frontier

```python
# Constraint: Sum of weights must equal 1
constraints = [
    {'type': 'eq', 'fun': lambda w: np.sum(w) - 1}  # Sum of weights = 1
]

# Objective function: Portfolio variance (to minimize)
def portfolio_variance(weights):
    weighted_returns = np.dot(monthly_returns, weights)  # Weighted monthly returns
    monthly_variance = np.var(weighted_returns)  # Variance of weighted returns
    annual_variance = monthly_variance * 12  # Annualize the variance
    return annual_variance

# Optimize to find the minimum variance portfolio
result = minimize(portfolio_variance, initial_weights, bounds=bounds, constraints=constraints)

# Extract results directly
optimal_weights = result.x
optimal_standard_deviation = portfolio_variance(optimal_weights) ** 0.5

# Create a DataFrame to show the weights with asset names
asset_names = basic_stats_df.loc[basic_stats_df.index != 'S&P500'].index  # Exclude 'S&P500'
weights_df = pd.DataFrame({'Asset': asset_names, 'Weight': optimal_weights})
weights_df.set_index('Asset', inplace=True)

# Display results
print("Optimal Weights:")
print(weights_df)
print(f"Global Minimum Portfolio Standard Deviation (Annualized): {optimal_standard_deviation:.6f}")

# Calculate the portfolio return for the optimal weights
optimal_return = np.dot(optimal_weights, returns)
print(f"Portfolio Return for Global Minimum Variance Portfolio: {optimal_return:.6f}")
```

```python
# Find the row with the closest 'Portfolio Standard Deviation' to optimal_standard_deviation
closest_std_index = (portfolio_sharpe_ratios_df['Portfolio Standard Deviation'] - optimal_standard_deviation).abs().idxmin()

# Filter the DataFrame to include rows from the closest standard deviation onwards
global_minimum_efficient_frontier = portfolio_sharpe_ratios_df.loc[closest_std_index:]

global_minimum_efficient_frontier
```

```python
# Plot the Minimum Variance Frontier
plt.figure(figsize=(10, 6))
plt.plot(
    global_minimum_efficient_frontier['Portfolio Standard Deviation'],  # X-axis: Standard Deviation (Risk)
    global_minimum_efficient_frontier['Target Return'],                # Y-axis: Target Return (Expected Return)
    label='Minimum Variance Frontier',
)
plt.xlabel('Portfolio Standard Deviation')
plt.ylabel('Expected Return')
plt.title('Global Minimum Variance Frontier')
plt.grid(True)
plt.legend()
plt.show()
```
