# Simple Return

```python
simple_returns = (data / data.shift(1))-1
simple_returns
```

### Remove NaNs

```python
# Exclude the first row
simple_returns = simple_returns.iloc[1:]
simple_returns
```

```python
# Count of NaN values in each column
nan_columns = simple_returns.isnull().sum()

# Filter columns with NaN count greater than 0
nan_columns = nan_columns[nan_columns > 0]

nan_columns
```

```python
# Drop columns with NaN values
# .drop(columns=...) droprops the columns listed
# .index retrieves the names of those columns
simple_returns = simple_returns.drop(columns=nan_columns.index)
simple_returns
```

```python
# Redefine the tickers list by removing the columns with NaN values
tickers = [ticker for ticker in tickers if ticker not in nan_columns.index]
tickers
```

```python
len(tickers)
```

### Risk Free Rate

```python
# ^IRX = 13-Week Treasury Bill Yield
risk_free_annual = yf.download('^IRX')['Adj Close'].iloc[-1] / 100
risk_free_annual
```

```python
risk_free_scaled = risk_free_annual / scaling_factor
risk_free_scaled
```

### Market Return

```python
market_data = yf.download('^GSPC', period='70y')['Adj Close']
market_data
```

```python
market_return = (market_data / market_data.shift(1))-1
market_return
```

```python
market_return_annual = market_return.mean() * 250
market_return_annual
```

### Market Risk Premium

```python
market_risk_premium = market_return_annual - risk_free_annual
market_risk_premium
```

### Excess Returns

```python
# Subtract risk_free_scaled from every numerical column in the DataFrame
excess_returns = simple_returns - risk_free_scaled

# Check the updated DataFrame
excess_returns
```

# Linear Regression

```python
from scipy import stats
import statsmodels.api as sm

# Initialize an empty dictionary to store the regression results
regression = {}

# Define X as the returns of ^GSPC
X = excess_returns['^GSPC']

# Add a constant for the intercept in statsmodels
X_with_const = sm.add_constant(X)

# Loop through each ticker in the tickers list
for ticker in tickers:
    Y = excess_returns[ticker]  # Get the returns of the current ticker

    # Perform linear regression with scipy
    slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)

    # Perform regression with statsmodels to get the p-value for the intercept
    model = sm.OLS(Y, X_with_const).fit()
    p_value_intercept = model.pvalues['const']  # Get p-value for intercept

    # Apply transformations
    alpha = intercept * scaling_factor  # Scale intercept (alpha) by one year
    r_square = abs(r_value) ** 0.5  # Take the square root of the absolute value of r_value

    # Store the transformed results in the dictionary
    regression[ticker] = {
        'slope': slope,
        'alpha': alpha,
        'r_square': r_square,
        'p_value_slope': p_value,  # P-value for slope
        'p_value_intercept': p_value_intercept,  # P-value for intercept
        'std_err': std_err,
    }

# Display the results
regression
```

### Alphas

```python
# Initialize an empty dictionary to store filtered results
filtered_regression = {}

# Loop through each ticker and its regression values
for ticker, values in regression.items():
    # Extract the p-value for the intercept
    p_value = values['p_value_intercept']  # Use the correct key for p-value of intercept
    alpha = values['alpha']  # Extract the alpha value

    # Check if the p-value is <= 0.1 and alpha > 0
    if p_value <= 0.1 and alpha > 0:
        filtered_regression[ticker] = values  # Add the ticker to the filtered dictionary

# Output filtered results
filtered_regression
```

```python
len(filtered_regression)
```

```python
# Sort the dictionary by the 'alpha' value in descending order
sorted_regression_alpha = dict(sorted(filtered_regression.items(), 
                                      key=lambda item: item[1]['alpha'], 
                                      reverse=True))

# Print the sorted dictionary
print("Sorted regression by intercept (higher to lower):")
sorted_regression_alpha
```

```python
top_20_alpha = list(sorted_regression_alpha.items())[:20]
top_20_alpha
```

### S&P Variance

```python
SP500_var = excess_returns['^GSPC'].var() * scaling_factor
SP500_var
```

### S&P Standard Deviation

```python
SP500_std = SP500_var ** 0.5
SP500_std 
```

### S&P Sharpe Ratio

```python
SP500_sharpe_r = market_risk_premium / SP500_std 
SP500_sharpe_r
```

### Variances, Betas, Systematic and Unsystematic Risks

```python
from scipy import stats

# Initialize the dictionary to store results
statistics = {}

# Loop through alpha tickers
for ticker in alpha_tickers:
    var = excess_returns[ticker].var() * scaling_factor  # Annual variance
    beta = sorted_regression_alpha[ticker]['slope']  # Beta
    sys_risk = beta**2 * SP500_var  # Systematic risk
    unsys_risk = var - sys_risk  # Unsystematic risk

    # Store results with labeled keys
    statistics[ticker] = {
        'Variance': var,
        'Beta': beta,
        'Systematic Risk': sys_risk,
        'Unsystematic Risk': unsys_risk
    }

# Display the results
statistics
```

```python
# Convert the statistics dictionary to a DataFrame
statistics_df = pd.DataFrame.from_dict(statistics, orient='index').T

# Display the resulting DataFrame
statistics_df
```

# Active Portfolio

```python
# Extract alpha values (item[1][1]) from top_20_alpha
alpha_values = {item[0]: item[1]['alpha'] for item in top_20_alpha}

# Initialize an empty dictionary
adjusted_alpha_values = {}

# Loop through each key-value pair in alpha_values to adjust apha at 2%
for key, value in alpha_values.items():
    # Multiply the value by 0.1 to keep 10%
    adjusted_value = value * 0.1
    
    # Add the key and adjusted value to the new dictionary
    adjusted_alpha_values[key] = adjusted_value

# Convert adjusted alpha values to a DataFrame row
alpha_row = pd.DataFrame([adjusted_alpha_values], index=["Alpha"])

# Concatenate the adjusted alpha row to statistics_df
statistics_df = pd.concat([statistics_df, alpha_row])

statistics_df
```

```python
# Calculate Reward to Risk Ratio
statistics_df.loc["Reward to Risk Ratio"] = statistics_df.loc["Alpha"] / statistics_df.loc["Unsystematic Risk"]
reward_to_risk_total = statistics_df.loc["Reward to Risk Ratio"].sum()
statistics_df.loc["Reward to Risk Ratio", "Total"] = reward_to_risk_total

# Calculate Active Weightings
statistics_df.loc["Active Weightings"] = statistics_df.loc["Reward to Risk Ratio"] / reward_to_risk_total

# Calculate Active Portfolio αA
statistics_df.loc["Active Portfolio αA"] = statistics_df.loc["Alpha"] * statistics_df.loc["Active Weightings"]
active_portfolio_total = statistics_df.loc["Active Portfolio αA"].sum()
statistics_df.loc["Active Portfolio αA", "Total"] = active_portfolio_total

# Calculate Portfolio Unsyst. Risk
statistics_df.loc["Portfolio Unsyst. Risk"] = statistics_df.loc["Unsystematic Risk"] * statistics_df.loc["Active Weightings"] ** 2
portfolio_unsyst_risk_total = statistics_df.loc["Portfolio Unsyst. Risk"].sum()
statistics_df.loc["Portfolio Unsyst. Risk", "Total"] = portfolio_unsyst_risk_total

# Calculate Active Portfolio Reward to Risk (αA/σεA²)
active_portfolio_reward_to_risk = active_portfolio_total / portfolio_unsyst_risk_total
statistics_df.loc["Active Portfolio Reward to Risk (αA/σεA²)", "Total"] = active_portfolio_reward_to_risk

# Calculate Annual information ratio
statistics_df.loc["Annual Information Ratio"] = statistics_df.loc["Active Portfolio αA"] / statistics_df.loc["Portfolio Unsyst. Risk"] ** 0.5

# Display the final DataFrame
statistics_df
```

### Active Portfolio βA

```python
active_portfolio_beta = (statistics_df.loc["Beta"] * statistics_df.loc["Active Weightings"]).sum()
active_portfolio_beta
```

### Market Reward to Risk

```python
market_reward_to_risk = market_risk_premium / SP500_var
market_reward_to_risk
```

### Weighting in Active Portfolio w0A

```python
weighting_in_active_portfolio = statistics_df.loc["Active Portfolio Reward to Risk (αA/σεA²)", "Total"] / market_reward_to_risk
weighting_in_active_portfolio
```

### Weight in Active Portfolio w*A (adjusting for beta)

```python
weighting_in_active_portfolio = weighting_in_active_portfolio / (1+(1 - active_portfolio_beta) * weighting_in_active_portfolio)
weighting_in_active_portfolio
```

### Weight in S&P500

```python
weighting_in_SP500 = 1 - weighting_in_active_portfolio
weighting_in_SP500
```

### Weights Without Leverage

```python
# Calculate the total sum of absolute weights
total_abs_weights = abs(weighting_in_SP500) + abs(weighting_in_active_portfolio)

# Normalize each weight
weighting_in_SP500_without_leverage = weighting_in_SP500 / total_abs_weights
weighting_in_active_portfolio_without_leverage = weighting_in_active_portfolio / total_abs_weights

# Print results
print("weighting_in_SP500_without_leverage:", weighting_in_SP500_without_leverage)
print("weighting_in_active_portfolio_without_leverage:", weighting_in_active_portfolio_without_leverage)
```

# Active Portfolio Results

### Final Weightings 

```python
final_weighting = weighting_in_active_portfolio * statistics_df.loc["Active Weightings"]
final_weighting_df = final_weighting.to_frame(name="Weightings").T
final_weighting_df
```

### Active Portfolio Information Ratio (αA/σεA)

```python
active_portfolio_information_ratio = statistics_df.loc["Active Portfolio αA", "Total"] / statistics_df.loc["Portfolio Unsyst. Risk", "Total"] ** 0.5
active_portfolio_information_ratio
```

### Revised Sharpe Ratio

```python
revised_sharpe_ratio = (SP500_sharpe_r  ** 2 + active_portfolio_information_ratio ** 2) ** 0.5
revised_sharpe_ratio = revised_sharpe_ratio.round(7)
revised_sharpe_ratio
```

### Sharpe Ratio Improvement

```python
sharpe_ratio_improvement = revised_sharpe_ratio - SP500_sharpe_r
sharpe_ratio_improvement = sharpe_ratio_improvement.round(7)
sharpe_ratio_improvement
```

# Verify Results

### Portfolio (P*) Risk Premium

```python
portfolio_risk_premium = weighting_in_SP500 * market_risk_premium + weighting_in_active_portfolio * (active_portfolio_beta * market_risk_premium + active_portfolio_total)
portfolio_risk_premium
```

### Active Portfolio Standard Deviation

```python
active_portfolio_std = (((active_portfolio_beta ** 2) * SP500_var) + portfolio_unsyst_risk_total) ** 0.5
active_portfolio_std
```

### Active Portfolio Covariance with M

```python
active_portfolio_cov_with_market = active_portfolio_beta * SP500_var
active_portfolio_cov_with_market
```

### Portfolio (P*) Standard Deviation

```python
portfolio_std = ((weighting_in_SP500 ** 2) * SP500_var + (weighting_in_active_portfolio ** 2) * (active_portfolio_std ** 2) + (2 * weighting_in_active_portfolio) * weighting_in_SP500 * active_portfolio_cov_with_market) ** 0.5
portfolio_std
```

### Portfolio (P*) Sharpe Ratio

```python
portfolio_sharpe_ratio = portfolio_risk_premium / portfolio_std
portfolio_sharpe_ratio = portfolio_sharpe_ratio.round(7)
print (portfolio_sharpe_ratio)
portfolio_sharpe_ratio > SP500_sharpe_r and portfolio_sharpe_ratio == revised_sharpe_ratio
```
